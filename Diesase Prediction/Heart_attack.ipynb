```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = ['Sex','Blood Pressure', 'Diet', 'Country', 'Continent', 'Hemisphere']
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])
df
```

```python
import pandas as pd
class_counts = df['Heart Attack Risk'].value_counts()
print("Class distribution:")
print(class_counts)
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd
X = df.drop(columns=['Heart Attack Risk'])
y = df['Heart Attack Risk']
smote = SMOTE(sampling_strategy='auto', random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)
print(pd.Series(y_resampled).value_counts())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_resampled,y_resampled,test_size=0.30,random_state=12)
len(X_train),len(X_test),len(y_train),len(y_test)
```

# Knn Classifier

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.utils import resample
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import numpy as np
```

```python
knn_classifier = KNeighborsClassifier()               #creating pipeline
scaler = StandardScaler()
pipeline = Pipeline([('scaler', scaler), ('knn', knn_classifier)])
```

```python
k_values = list(range(3, 10))
from sklearn.model_selection import cross_val_score
cv_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=15, scoring='accuracy')  # 15-fold cross-validation
    cv_scores.append(scores.mean())

optimal_k = k_values[np.argmax(cv_scores)]
print(f"Optimal K: {optimal_k}")

for k, score in zip(k_values, cv_scores):
    print(f"K = {k}, Accuracy = {score:.4f}")

plt.figure(figsize=(10, 6))
plt.plot(k_values, cv_scores, marker='o', linestyle='-', color='b')
plt.title('Error Rate vs. K Value')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```
